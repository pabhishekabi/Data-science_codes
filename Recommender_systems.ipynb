```python
user_movies_df=movies_df.pivot(index='userId',
                              columns='movie',
                               values='rating'). reset_index(drop=True)
```

```python
user_movies_df
```

```python
user_movies_df.index=movies_df.userId.unique()
```

```python
user_movies_df
```

```python
user_movies_df.fillna(0, inplace=True)
```

```python
user_movies_df
```

```python
#calculating cosine similarity between users 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sin = 1- pairwise_distances(user_movies_df.values,metric='cosine')
```

```python
user_sin
```

```python
#store the results in dataframe 
user_sin_df=pd.DataFrame(user_sin)
```

```python
#SET THE INDEX AND COLUMN NAMES TO USER ID 
user_sin_df.index=movies_df.userId.unique()
user_sin_df.columns=movies_df.userId.unique()
```
